# Calcul scientifique avec Python


```python
# import des modules de bases
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

# config taille images, en inches
plt.rcParams["figure.figsize"] = (10, 10)
```

On va définir l'intervalle $[t_{min}, \dots, t_{max}]$ considéré et le pas entre deux $t_i$.

```python
# pas du temps
T_MIN = -4
T_MAX = 4
NB_STEPS = 1_000
DELTA_T = (T_MAX - T_MIN) / NB_STEPS
ts = np.linspace(T_MIN, T_MAX, num=NB_STEPS)
```

On définit la vitesse comme paramétrée par $t$ avec $vx(t) = \cos(t^2)$ et $vy(t) = \sin(t^2)$.

```python
# fonctions vitesses
vx = lambda t: np.cos(t**2)
vy = lambda t: np.sin(t**2)
```

On a besoin d'intégrer $x(t) = \int_0^t vx(u) du$ et  $y(t) = \int_0^t vy(u) du$ pour $t \in [t_{min}, \dots, t_{max}]$.
Plutôt que de recalculer $\int_0^t vx(u) du$ pour tous les temps, on peut faire l'intégrale _cumulative_ (qui fait la somme successives pour les $t$ croissants).

```python

# intégrale cumulative
ys = integrate.cumulative_trapezoid(vy(ts), ts)
xs = integrate.cumulative_trapezoid(vx(ts), ts)

# plus élégant et rapide que recalculer
# x = lambda t: integrate.quad(vx, 0, t)
# y = lambda t: integrate.quad(vy, 0, t)
```

Plus qu'à tracer les couples $(y, x)$ et afficher.

```python
plt.scatter(xs, ys, marker="+", s=1, c=None)
plt.show()
```
